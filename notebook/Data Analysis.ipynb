```python
num_plots = len(merged_df.Crop.unique())
cols = 2 # Define the number of columns
rows = (num_plots // cols) + (num_plots % cols > 0)  # Calculate the required rows

fig, axes = plt.subplots(rows, cols, figsize=(15, 30), constrained_layout=True)
axes = axes.flatten()

# Plot each crop's data in a separate subplot
for i, (crop, crop_data) in enumerate(merged_df.groupby('Crop')):
    if i >= num_plots:
        break  # Stop if we exceed the number of required subplots
    
    sns.lineplot(data=crop_data, x='year', y='Production (Tonnes)', ax=axes[i], marker='o')
    axes[i].set_title(crop)
    axes[i].set_xlabel('Year')
    axes[i].set_ylabel('Production')

# Turn off any remaining empty subplots
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

plt.suptitle('Year-wise Production for Each Crop', fontsize=12)
plt.show()
```
